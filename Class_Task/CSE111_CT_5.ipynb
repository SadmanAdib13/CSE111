```python
#Task 1
class Student:
    def __init__(self, name, student_id, program='unknown'):
        self.name = name
        self.student_id = student_id
        self.program = program

    def dailyEffort(self, hours):
        self.daily_effort = hours

    def printDetails(self):
        print(f"Name: {self.name}")
        print(f"Student ID: {self.student_id}")
        print(f"Program: {self.program}")
        print(f"Daily Effort: {self.daily_effort} hours")

        if self.daily_effort<=4:
            print("Suggestion: Keep up the good work!")
        elif self.daily_effort<=2:
            print("Suggestion: Should give more effort!")
        else:
            print("Suggestion: Excellent! Now motivate others.")

# Creating instances of the Student class
harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')

john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')

naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
class Sphere:
    def __init__(self, n, r=1, c="White"):
        self.name = n
        self.radius = r
        self.color = c
        self.volume = (4 / 3) * 3.1416 * (self.radius ** 3)

    def printDetails(self):
        print(f"Sphere ID: {self.name}")
        print(f"Color: {self.color}")
        print(f"Volume: {self.volume}")

    def merge_sphere(self, *args):
        print(f"Spheres are being merged")
        for i in args:
            self.volume = self.volume + i.volume
            if self.color != i.color:
                self.color = "Mixed Color"

# Driver code
sphere1 = Sphere("Sphere 1")
print("1***************")
sphere1.printDetails()

sphere2 = Sphere("Sphere 2", 3)
print("2***************")
sphere2.printDetails()

sphere3 = Sphere("Sphere 3", 2)
print("3***************")
sphere3.printDetails()

sphere3.merge_sphere(sphere1, sphere2)
print("4***************")
sphere3.printDetails()

sphere4 = Sphere("Sphere 4", 5, "Purple")
print("5***************")
sphere4.merge_sphere(sphere3)
print("6***************")
sphere4.printDetails()
```
